```python
# Calculate risk-preference

# p: reward prob 
p = filtered_T['risk_prob']
# r: reward mag
filtered_T.loc[filtered_T['chosen'] == 0, 'r'] = 0

# Filter rows where 'chosen' is 1
filtered_T.loc[filtered_T['chosen'] == 1, 'r'] = filtered_T['risk_gain'] - filtered_T['sure_gain']
r = filtered_T['r']



# Calculate risk-preference
filtered_T.insert(0,'expected value E(X)', p * r) 
filtered_T.insert(0,'risk of the option V(x)', (r**2) * p*(1-p))
```
